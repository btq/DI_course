# Project Structure

If you're doing anything but the simplest one-off exploratory analysis, it turns out to be extraordinarily helpful to make everything you do **easily repeatable**.  At its simplest, this consists of a few guidelines:

1. Do everything from a script.  (Or at least: If you perform an action, add it to a script so that you remember it later.)
2. Document your "pipeline": What order your scripts should be run in.  
   - Preferably document it in executable form (i.e., by a script of some sort);
   - .. and so that it can be re-started from the middle.
3. Keep your data in a separate folder, and write a script for getting it.
4. Put the scripts and the documentation into git.

There are at least three good reasons for this:

- You will realize you made a mistake in your data cleaning step, and that you need to re-run it and everything after it.
- You will realize that you need to run a slightly tweaked version of your process.
- You will want to run your process on a different machine.

Recording the pipeline: `make`
-------------------------------
`make` is a special purpose tool designed for managing dependencies.  The basic model is that you have a bunch of sections that look like

>     target_file: dependency_file1 dependency_file2....
>             some command to run
>             a command to run after it
>             ....

Then, when you type `make target_file` make will first check if all of the dependecy files are are up-to-date.  If they are not, it will first make _them_.  Once that's done, it will run the commands in order.

If you just run `make` without specifying a target, it will run the top one in the file.  It is customary for this to be something like

>     all: [all your 'top-level' target]

**Caveats and gotchas:**
  1. What does up-to-date mean?  `make` is based around the idea of actual _files_ (not just abstract targets) and so it will check if the file exists (if not, it has to be made) and if it does it will use its _filesystem timestamp_ to determine if it has changed since the last run.

  This means that if you want to force target_file to be remade, you could either delete it or run `touch dependency_file1` to update the timestamp on dependency_file1.

  2. It is white-space sensitive.  Worse, see those spaces before the commands: they are TABs, not spaces.  The use of fixed tabs has fallen out of favor lately, so it may be difficult to make your text-editor actually put a space rather than 4 spaces.  For instance, in IPython Notebook the easiest way is to copy-paste in a tab from elsewhere!  
  
  (In the code sample below, the little arrows are teling us that it is really a tab.)
  
  3. Nowadays, we often do things other than generating output files from input files.  For instance, we might write to the database.  Or we might perform an operation where we do not know -- a priori -- the names of the files it will generate (i.e., downloading some sequence of files).  
  
  `make` doesn't want to know.  The fix is to use a "sentinel file" -- an empty file whose presence (or absence, or timestamp) indicates whether our operation has been done.
  
  4. Make has a complicated variable substitution system.  Fortunately, it's not really necessary to know much of it.

The contents of the Makefile are roughly:
```
# This is a simple data pipeline:
#
# Ways in which it is simple:
# - it assumes local data storage (i.e., we never have to query a server to see if something's changed);
# - we do only simple dependencies between stages (e.g., we don't worry about fetching different parts of the same dataset in parallel)
#
# Ways in which it is not:
# - we use a database to store out data, so the targets aren't really files in the ordinary sense; as a result,
# - we use 'sentinel files' -- that is files with no content, but whose presence and timestamp indicate that some operation has been completed
#

# We're going to declare the command we use to talk to the database
PSQL_HMDA=psql --set ON_ERROR_STOP=1 hmda-lar
all: data/.make_mktshare_summary

# Make a summary of the FHA marketshare data
data/.make_mktshare_summary: data/.make_hmda_mktshare src/db_hmda/make_summary.sql
	cat src/db_hmda/make_summary.sql | ${PSQL_HMDA}
	touch data/.make_summary

# Do some processing on the HMDA data, to pull out FHA marketshare by MSA / census tract
data/.make_hmda_mktshare: data/.make_hmda_db src/db_hmda/analyze_geo_msa.sql src/db_hmda/analyze_geo_tract.sql
	cat src/db_hmda/analyze_geo_msa.sql | ${PSQL_HMDA}
	cat src/db_hmda/analyze_geo_tract.sql | ${PSQL_HMDA}
	touch data/.make_process

## Downloading HMDA data and loading it into the DB.  The data dictionaries (in dicts/) are hand-made.
data/.make_hmda_db: data/.make_hmda_dl src/db_hmda/make_tables.sql dicts/agencies.csv dicts/loan-types.csv src/db_hmda/python-na.py
	dropdb --if-exists hmda-lar
	createdb hmda-lar
	cat src/db_hmda/make_tables.sql | ${PSQL_HMDA}
	cat "data/raw/hmda/2012PMICLAR_-_National.zip" | funzip | src/db_hmda/python-na.py | ${PSQL_HMDA} -c "COPY PMIC_LAR FROM stdin WITH DELIMITER ',' CSV NULL AS '';"
	cat "data/raw/hmda/2012HMDALAR_-_National.zip" | funzip | src/db_hmda/python-na.py | ${PSQL_HMDA} -c "COPY HMDA_LAR FROM stdin WITH DELIMITER ',' CSV NULL AS '';"
	touch data/.make_hmda_db

data/.make_hmda_dl: src/get_hmda_2012.sh
	bash src/get_hmda_2012.sh
	curl http://www2.census.gov/geo/gazetteer/2013_Gazetteer/2013_Gaz_tracts_national.zip | funzip > data/raw/2013_Gaz_tracts_national.tsv
	touch data/.make_hmda_dl

### Cleanup...
clean:
	rm -r data/*
	rm -f data/.make_*

# If we need to
drop_hmda_db:
	dropdb --if-exists hmda-lar
	rm -f data/.make_hmda_db
```

   **A word on idempotence**
   
   The script for creating the table follows next.  There's one thing worth noting about it: 
   _We drop the table (if it exists) before creating it._ 
   
   This is so that the operation is **idempotent**.  This means that performing it twice is the same as performing it once.  This makes it much easier to reason about the state of the system or recover from failure.  Otherwise, trying to run it twice would give an error and break our pipeline.

And a useful cleaning script in `src/db_hmda/python-na.py` would look like:
```python
#!/usr/bin/python
import sys
for line in sys.stdin:
    def munge_na(x):
        s = x.strip()
        if s.startswith("NA"):
            return ""
        return s
    print ",".join(map(munge_na, line.split(",")))
```

We can preform some basic computation in SQL.  For instance in `src/db_hmda/analyze_geo_tract.sql` we might have the following code for aggregating by census tract:
```SQL
DROP TABLE IF EXISTS Tract_Level;
CREATE TABLE Tract_Level AS
    SELECT State_Code, County_Code, Census_Tract_Number, Loan_Type, Action_Type, COUNT(1) as NUM_BORROWERS, sum(Loan_Amount) as TOTAL_AMOUNT
    FROM HMDA_LAR
    WHERE Loan_Purpose = 1 AND Property_Type = 1
    GROUP BY State_Code, County_Code, Census_Tract_Number, Loan_Type, Action_Type;

DROP TABLE IF EXISTS FHA_by_tract;
CREATE TABLE FHA_by_tract AS
    SELECT State_Code, County_Code, Census_Tract_Number, SUM(NUM_BORROWERS) AS NUM_ALL, SUM(NUM_BORROWERS * (Loan_Type = 2)::int) AS NUM_FHA, SUM(TOTAL_AMOUNT) as AMT_ALL, SUM(TOTAL_AMOUNT * (Loan_Type = 2)::int) as AMT_FHA
    FROM Tract_Level
    WHERE ACTION_TYPE = 1
    GROUP BY State_Code, County_Code, Census_Tract_Number;

ALTER TABLE FHA_by_tract ADD PCT_NUM_FHA REAL;
UPDATE FHA_by_tract
    SET PCT_NUM_FHA = NUM_FHA*100.0/NUM_ALL;

ALTER TABLE FHA_by_tract ADD PCT_AMT_FHA REAL;
UPDATE FHA_by_tract
    SET PCT_AMT_FHA = AMT_FHA*100.0/AMT_ALL;

\COPY ( SELECT State_Code, County_Code, Census_Tract_Number, NUM_ALL, NUM_FHA, PCT_NUM_FHA, AMT_ALL, AMT_FHA, PCT_AMT_FHA  FROM FHA_by_tract ORDER BY PCT_AMT_FHA DESC ) TO 'data/fha_by_tract.csv' WITH CSV
```
